```python
def check_accuracy(y_pred, y_test):
    result = 0
    for i in range(len(y_pred)):
        if y_pred[i] != y_test[i]:
            result += 1
    score = 1 - (result/len(y_pred))
    return score

# Validation linear
model_validation_linear = svm.SVC(C=3, kernel='linear')
model_validation_linear.fit(X, y)
y_pred_linear = model_validation_linear.predict(X_test)
score_linear = check_accuracy(y_pred_linear, y_test)
print("Linear:", score_linear)

# Validation RBF
model_validation_rbf = svm.SVC(C=2, gamma=1, kernel='rbf')
model_validation_rbf.fit(X, y)
y_pred_rbf = model_validation_rbf.predict(X_test)
score_rbf = check_accuracy(y_pred_rbf, y_test)
print("RBF:", score_rbf)

#Validation Polynomial
model_validation_poly = svm.SVC(C=5, gamma=1, degree=3, kernel='poly')
model_validation_poly.fit(X, y)
y_pred_poly = model_validation_poly.predict(X_test)
score_poly = check_accuracy(y_pred_poly, y_test)
print("Poly:", score_poly)
```

```python
def plot_decision_boundary_contour(clf, X, Y, cmap='Paired_r'):
    h = 0.1
    x_min, x_max = X[:,0].min() - 10*h, X[:,0].max() + 10*h
    y_min, y_max = X[:,1].min() - 10*h, X[:,1].max() + 10*h
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(5,5))
    plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.25)
    plt.contour(xx, yy, Z, colors='k', linewidths=0.7)
    plt.scatter(X[:,0], X[:,1], c=Y, cmap=cmap, edgecolors='k');
```

### Plotting the decision boundary

#### Linear Model

```python
plot_decision_boundary_contour(model_validation_linear, X, y)
```

#### Rbf model

```python
plot_decision_boundary_contour(model_validation_rbf, X, y)
```

#### Polynomial Model

```python
plot_decision_boundary_contour(model_validation_poly, X, y)
```
