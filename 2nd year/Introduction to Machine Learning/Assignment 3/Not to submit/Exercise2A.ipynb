```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np
```

```python
data = np.loadtxt('artificial.csv', delimiter=',',dtype='float64')
print(data)
```

```python
X = np.loadtxt('artificial.csv', delimiter=',', dtype='float64', usecols=(0,1))
y = np.loadtxt('artificial.csv', delimiter=',', dtype='float64', usecols=2)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=3)
clf = clf.fit(X_train, y_train)
```

```python
predictions = clf.predict(X_test)
predictions
```

```python
clf.predict_proba(X_test)
```

```python
from sklearn import tree
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   filled=True,
                   fontsize=12)
```

```python
from matplotlib.colors import ListedColormap
h = 1
X_min, X_max = X[:, 0].min() - 1, X[:, 0].max() + 1
Y_min, Y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(X_min, X_max, h), np.arange(Y_min, Y_max, h))
x1, x2 = xx.ravel(), yy.ravel()
clf = DecisionTreeClassifier(max_depth=3)
XXe = np.c_[x1, x2]
clf = clf.fit(X_train, y_train)
p = clf.predict(XXe)
classes = p > 0.5
clz_mesh = classes.reshape(xx.shape)
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

plt.figure(2)
plt.pcolormesh(xx, yy, clz_mesh, cmap=cmap_light)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cmap_bold)
plt.show()
```
